```python
import requests
from lxml import html

# SCRAPE HOME PAGE OF KLARNA
home_page = "https://www.klarna.com/us/"

def get_source(page_url):
    r = requests.get(page_url)
    source = html.fromstring(r.content)

    return source

source = get_source(home_page)
source
```

```python
#  SCRAPE ALL CATEGORY LINKS AND ADDS TO A LIST

category_list = source.xpath("//a[contains (@class, 'sc-gtssRu gmjWml sc-hBMVcZ gRFIVK sc-dvXXZy gIJatV')]")
category_list_links = "@href"
category_name = ".//span[contains (@class, 'sc-gKAblj tbujR')]/text()"
all_category_links = []
categories = []

for l in category_list:
    links = l.xpath(category_list_links)
    names = l.xpath(category_name)

    if links != ['/us/shopping/']:
        val = str(links)
        new_val = val.replace("]", "")
        new2_val = new_val.replace("[", "")
        final_links = eval(new2_val)
        all_category_links.append(final_links)

    res = str(names)
    new_res = res.replace("]", "")
    new2_res = new_res.replace("[", "")
    final_names = eval(new2_res)
    categories.append(final_names)

if final_names != "Shopping categories":
    file = open("categories.txt", mode ="a")
    file.write(f"{categories}\n")
    file.close()



        
print(all_category_links)
print()
print(categories)


```

```python
# ITERATES THROUGH CATEGORY LIST AND SCRAPES ALL SUB-CATEGORY LINKS AND ADDS IN A NEW LIST

sub_categories_link = []
sub_categories_names = []

new_link = "//a[contains (@class, 'css-1v7kxj3')]"
sub_category_name = ".//h2[contains (@class, 'pUoKQGvtG9 vy0GbKu6J6 hENphQLjTi')]/text()"


for x in all_category_links:
    link_source = get_source(f"https://www.klarna.com{x}")
    sub_cat_link = link_source.xpath(new_link)


    
    for a in sub_cat_link:
        sub_links = a.xpath(category_list_links)
        names = a.xpath(sub_category_name)


        sub_val = str(sub_links)
        sub_new_val = sub_val.replace("]", "")
        sub_new2_val = sub_new_val.replace("[", "")
        sub_final_links = eval(sub_new2_val)
        sub_categories_link.append(sub_final_links)
        
        res = str(names)
        new_res = res.replace("]", "")
        new2_res = new_res.replace("[", "")
        final_names = new2_res.replace("'", "")
        sub_categories_names.append(final_names)

file = open("sub_categories.txt", mode ="w")
file.write(f"{sub_categories_names}\n")
file.close()
        
print(len(sub_categories_link))
print()
print(sub_categories_link)
print()
print(sub_categories_names)
    
```

```python
# ITERATES THROUGH SUB-CATEGORIES LIST AND SCRAPES ALL SUB-SUB-CATEGORY LINKS AND ADDS IN A NEW LIST

sub_sub_categories_link = []
sub_new_link = "//a[contains (@class, 'css-1a99541')]"

for x in sub_categories_link:
    link_source = get_source(f"https://www.klarna.com{x}")
    sub_cat_link = link_source.xpath(sub_new_link)
    
    for a in sub_cat_link:
        sub_links = a.xpath(category_list_links)
        sub_val = str(sub_links)
        sub_new_val = sub_val.replace("]", "")
        sub_new2_val = sub_new_val.replace("[", "")
        
        sub_final_links = eval(sub_new2_val)
        sub_sub_categories_link.append(sub_final_links)
        
print(len(sub_sub_categories_link))
print()
print(sub_sub_categories_link)
```

```python
# ITERATES THROUGH EACH SUB-SUB-CATEGORY LIST AND SCRAPES EACH PRODUCT LINK ADDING THEM TO A LIST

product_link = []
product_new_link = "//div[contains (@class, 'k6oEmfY83J css-6r3upd')]"

for x in sub_sub_categories_link:
    link_source = get_source(f"https://www.klarna.com{x}")
    product_links = link_source.xpath(product_new_link)

    for a in product_links:
        product_sub_links = a.xpath(".//a/@href")
        product_sub_val = str(product_sub_links)
        product_sub_new_val = product_sub_val.replace("]", "")
        product_sub_new2_val = product_sub_new_val.replace("[", "")
        product_link.append(product_sub_new2_val)

        
print(len(product_link))
print()
print(product_link)
```

```python
# ITERATES THROUGH THE PRODUCTS LINK LIST AND SCRAPES THE NAME OF EACH PRODUCT AND ADDS THEM TO A LIST

product_names = []
get_product_name = "//div[contains (@class, 'o0zhtk3chz')]" 

for x in product_link:
    new = x.replace("'", "")
    link_source = get_source(f"https://www.klarna.com{new}")
    retrieve_product_name = link_source.xpath(get_product_name)

    for a in retrieve_product_name:
        product_sub_links_name = a.xpath(".//h1[contains (@class, 'pUoKQGvtG9 GbZ_OCIf9m fNW2u2e0aB nsNMYyHYau INPZTLX0wh')]/text()")
        product_sub_val_name = str(product_sub_links_name)
        product_sub_new_val_name = product_sub_val_name.replace("]", "")
        product_sub_new2_val_name = product_sub_new_val_name.replace("[", "")
        product_names.append(product_sub_new2_val_name)
        

        
print(len(product_names))
print()
print(product_names)
```
