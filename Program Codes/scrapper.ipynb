```python
import requests
from lxml import html

# SCRAPE HOME PAGE OF KLARNA
home_page = "https://www.klarna.com/us/"

def get_source(page_url):
    r = requests.get(page_url)
    source = html.fromstring(r.content)

    return source

source = get_source(home_page)
source
```

```python
#  SCRAPE ALL CATEGORY LINKS AND ADDS TO A LIST

category_list = source.xpath("//a[contains (@class, 'sc-gtssRu gmjWml sc-hBMVcZ gRFIVK sc-dvXXZy gIJatV')]")
category_list_links = "@href"
category_name = ".//span[contains (@class, 'sc-gKAblj tbujR')]/text()"
all_category_links = []
categories = []

for l in category_list:
    links = l.xpath(category_list_links)
    names = l.xpath(category_name)

    if links != ['/us/shopping/']:
        val = str(links)
        new_val = val.replace("]", "")
        new2_val = new_val.replace("[", "")
        final_links = eval(new2_val)
        all_category_links.append(final_links)

    res = str(names)
    new_res = res.replace("]", "")
    new2_res = new_res.replace("[", "")
    final_names = eval(new2_res)
    categories.append(final_names)

if final_names != "Shopping categories":
    file = open("../TXT files/categories.txt", mode ="w")
    file.write(f"{categories}\n")
    file.close()
print(categories)
```

```python
# ITERATES THROUGH CATEGORY LIST AND SCRAPES ALL SUB-CATEGORY LINKS AND ADDS IN A NEW LIST

sub_categories_link = []
sub_categories_names = []

new_link = "//div[contains (@class, 'KlZPLvU93Y pr-1u8qly9')]/a"
sub_category_name = ".//h2[contains (@class, 'pUoKQGvtG9 vy0GbKu6J6 hENphQLjTi pr-0')]/text()"


for x in all_category_links:
    link_source = get_source(f"https://www.klarna.com{x}")
    sub_cat_link = link_source.xpath(new_link)

    for a in sub_cat_link:
        sub_links = a.xpath(category_list_links)
        names = a.xpath(sub_category_name)


        sub_val = str(sub_links)
        sub_new_val = sub_val.replace("]", "")
        sub_new2_val = sub_new_val.replace("[", "")
        sub_final_links = eval(sub_new2_val)
        sub_categories_link.append(sub_final_links)
        
        res = str(names)
        new_res = res.replace("]", "")
        new2_res = new_res.replace("[", "")
        final_names = new2_res.replace("'", "")
        sub_categories_names.append(final_names)
        

file = open("../TXT files/sub_categories.txt", mode ="w")
file.write(f"{sub_categories_names}\n")
file.close()

```

```python
#  SCRIPT TO FILTER SPECIFIC PRODUCT INFORMATION FROM API RESPONSE
import json

section = []
specifications = []

for x in items:

    p = x["product"]
    product_url = p["url"]
    images = p["images"]
    product_id = p["id"]
    product_name = p["name"]
    description = p["description"]

    
    b = x['brand']
    if b == None:
        brand_id = "null"
        brand_name = "null"
    else:
        brand_id = b['id']
        brand_name = b['name']
    
    min_price = x['prioritizedMinPrice']
    max_price = x['prioritizedMaxPrice']

    s = x["specification"]

    spec = s['sections']

    created_time = s['creationTime']

    rating = x['rating']

    bread_crums = x["breadcrumbs"]

    for item in bread_crums:
        cat = item["name"]
        section.append(cat)

    category = section[0]
    sub_category = section[1]
    sub_sub_category = section[2]

    for x in spec:
        sec = x["attributes"]
        specifications.clear()

        for v in sec:
            for key , val in v.items():
                if key == "name":
                    z = val
                elif key == "values":
                    t = val[0]
                    q = t["name"]
                    all = {z:q}
                    specifications.append(all)
        
        
    final_data = {"images":images, "product_id": product_id, "product_name": product_name, "description": description,  "product_url": product_url, "brand_id":brand_id, "brand_name": brand_name, "specifications": specifications, "created_time": created_time, "rating": rating, "bread_crums": f"{category} & {sub_category} & {sub_sub_category}", "min_price": min_price, "max_price": max_price}


    print(final_data)
    res = json.dumps(final_data)
    file = open("KalrnaData.json", mode = "a")
    file.write(f"\n {res}")
    file.close()

```
